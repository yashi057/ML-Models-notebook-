```python
import tkinter as tk
import nltk
from nltk.chat.util import Chat, reflections
import pyttsx3

engine=pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()



# Download once
nltk.download('punkt')

# Define chatbot patterns
pairs = [
    ["hi|hello", ["Hello 👋", "Hi there!"]],
    ["what is your name ?", ["I am a chatbot"]],
    ["how are you ?", ["I am fine 😊"]],
    ["(.*) your name ?", ["My name is ChatBot"]],
    ["bye", ["Goodbye!", "See you later"]],
    ["(.*)", ["Sorry, I don't understand 😅"]]
]

chatbot = Chat(pairs, reflections)

# Function for GUI
def send():
    user_input = entry.get()
    chat_area.insert(tk.END, "You: " + user_input + "\n")

    response = chatbot.respond(user_input)
    if response is None:
        response = "Sorry, I didn't understand."

    chat_area.insert(tk.END, "Bot: " + response + "\n")
    speak(response)
    entry.delete(0, tk.END)

# GUI window
root = tk.Tk()
root.title("NLTK Chatbot")

chat_area = tk.Text(root, height=20, width=50)
chat_area.pack()

entry = tk.Entry(root, width=40)
entry.pack()

send_button = tk.Button(root, text="Send", command=send)
send_button.pack()

# Enter key support
entry.bind("<Return>", lambda event: send())

root.mainloop()
```

```python
import tkinter as tk
import pyttsx3
import speech_recognition as sr
import webbrowser
import datetime
import threading

# ---------------- Voice Engine ----------------
engine = pyttsx3.init()
engine.setProperty('rate', 170)

def speak(text):
    chat_area.insert(tk.END, "Jarvis: " + text + "\n")
    chat_area.see(tk.END)
    engine.say(text)
    engine.runAndWait()

# ---------------- Voice Input FIXED ----------------
def take_command():
    r = sr.Recognizer()

    try:
        with sr.Microphone() as source:
            chat_area.insert(tk.END, "🎤 Listening...\n")
            chat_area.see(tk.END)

            r.adjust_for_ambient_noise(source, duration=1)
            r.pause_threshold = 1

            audio = r.listen(source, timeout=5, phrase_time_limit=6)

        query = r.recognize_google(audio, language="en-in")
        chat_area.insert(tk.END, "You (voice): " + query + "\n")
        chat_area.see(tk.END)
        return query.lower()

    except sr.WaitTimeoutError:
        speak("No voice detected")
        return ""

    except sr.UnknownValueError:
        speak("Sorry, I did not understand")
        return ""

    except sr.RequestError:
        speak("Network error")
        return ""

    except Exception as e:
        speak("Microphone error")
        return ""

# ---------------- Command System ----------------
def process_command(query):

    if query == "":
        return

    # Greetings
    if "hello" in query or "hi" in query:
        speak("Hello! I am Jarvis.")

    # Time
    elif "time" in query:
        time = datetime.datetime.now().strftime("%H:%M:%S")
        speak("Current time is " + time)

    # Date
    elif "date" in query:
        date = datetime.datetime.now().strftime("%d %B %Y")
        speak("Today is " + date)

    # YouTube
    elif "youtube" in query:
        webbrowser.open("https://youtube.com")
        speak("Opening YouTube")

    # Google
    elif "google" in query:
        webbrowser.open("https://google.com")
        speak("Opening Google")

    # Search
    elif "search" in query:
        query = query.replace("search", "")
        webbrowser.open("https://www.google.com/search?q=" + query)
        speak("Searching Google")

    # Exit
    elif "exit" in query or "bye" in query:
        speak("Goodbye!")
        root.quit()

    else:
        speak("I am still learning this command")

# ---------------- Button Functions (THREAD SAFE) ----------------
def send_text():
    query = entry.get().lower()
    chat_area.insert(tk.END, "You: " + query + "\n")
    chat_area.see(tk.END)
    entry.delete(0, tk.END)
    process_command(query)

def voice_input():
    threading.Thread(target=lambda: process_command(take_command())).start()

# ---------------- GUI ----------------
root = tk.Tk()
root.title("JARVIS AI Assistant")
root.geometry("450x550")

chat_area = tk.Text(root, height=25, width=55)
chat_area.pack(pady=10)

entry = tk.Entry(root, width=40)
entry.pack(pady=5)

btn_frame = tk.Frame(root)
btn_frame.pack()

send_btn = tk.Button(btn_frame, text="Send", command=send_text)
send_btn.grid(row=0, column=0, padx=5)

voice_btn = tk.Button(btn_frame, text="🎤 Speak", command=voice_input)
voice_btn.grid(row=0, column=1, padx=5)

root.mainloop()
```
